```python
#Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
```

```python
# Create a plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
#Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
#Create a quick plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
#Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth = 2, label = 'Boston')
#Create labels for x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Set the y limit between 0 and 45.
plt.ylim(0, 45)
#Create a title.
plt.title("PyBer Fare by Month")
#Add grid
plt.grid()
#Add a legend
plt.legend()
#Change the size
plt.figure(figsize=(20,10))
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker = 'D', color = 'green', label = 'Boston')
ax.legend()
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, color = 'red', label = 'Chicago')
plt.legend()
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.xlim(0,50)
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#using list comprehension
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c = 'skyblue', linewidth = 2, alpha = 0.8, edgecolors= 'black', s = [i * 5 for i in y_axis], label = 'Boston')
ax.legend()
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.set_title('PyBer Fare by Month')
ax.set_xlim(0,50)
ax.invert_yaxis()
```

```python
plt.pie(y_axis, labels= x_axis)
plt.show()
```

```python
#Assign 12 colors, one for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen","greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize = (8,8))
plt.pie(y_axis, labels = x_axis, colors = colors, explode = explode_values, autopct = '%.1f%%')
plt.gca().invert_xaxis()
plt.show()
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen","greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explodes = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis, labels = x_axis, labeldistance= 1.2, colors = colors, startangle = 90, shadow = True, explode = explodes, autopct ='%.1f%%')
ax.invert_xaxis()
```
